```python
from matplotlib.legend_handler import HandlerPatch
import matplotlib.patches as mpatches
```

```python
class HandlerEllipse(HandlerPatch):
    def create_artists(self, legend, orig_handle,
                       xdescent, ydescent, width, height, fontsize, trans):
        center = 0.5 * width - 0.5 * xdescent, 0.5 * height - 0.5 * ydescent
        p = mpatches.Ellipse(xy=center, width=orig_handle.width,
                                        height=orig_handle.height)
        self.update_prop(p, orig_handle, legend)
        p.set_transform(trans)
        return [p]
    #from https://stackoverflow.com/questions/65631607/showing-different-size-circles-in-heatmap-with-legend-using-matplotlib/65654470
```

```python
# adapted from https://stackoverflow.com/questions/59381273/heatmap-with-circles-indicating-size-of-population
x, y = np.meshgrid(np.arange(c.shape[1]), np.arange(c.shape[0]))
fig, ax = plt.subplots()
circles = [plt.Circle((j,i), radius=R * 0.25) for R,j, i in zip(c2.flat, x.flat, y.flat)]
col = PatchCollection(circles, array=c.flatten(), cmap=cmap, norm=matplotlib.colors.Normalize(vmin=-1, vmax=1))
#col.set_clim(vmax = 6, vmin = -4)
ax.add_collection(col)


ax.set(xticks=np.arange(c.shape[1]), yticks=np.arange(c.shape[0]),
       xticklabels=cc.columns, yticklabels=cc.index,)
ax.set_xticks(np.arange(c.shape[1]+1)-0.5, minor=True)
ax.set_yticks(np.arange(c.shape[0]+1)-0.5, minor=True)

plt.grid(which='major', color='black')
ax.grid(visible = False)
ax.set_aspect('equal',adjustable="box")
ax.spines['left'].set_color('0.5')
ax.spines['bottom'].set_color('0.5')
ax.set_facecolor((1, 1, 1))
ax.tick_params(bottom=True,
               left=True,color = 'black'
              )
fig.colorbar(col)

plt.ylabel('Identity')
plt.xlabel('Motifs')
plt.xticks(rotation=35, ha='right')
texts = ["10","50","75"]
c = [mpatches.Ellipse((), width=10*0.25, height=10*0.25, color="grey"),
     mpatches.Ellipse((), width=50*0.25, height=50*0.25, color="grey"),
     mpatches.Ellipse((), width=75*0.25, height=75*0.25, color="grey"),
    ]
legend = ax.legend(c,texts, handler_map={mpatches.Ellipse: HandlerEllipse()},title="Percent Expressed",bbox_to_anchor=(3.50, 0.82, 1.0, .102),fontsize="large")
plt.setp(legend.get_title(),fontsize='large')

figg = plt.gca()
```
